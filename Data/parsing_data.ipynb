```python
import pandas as pd
import json

# Import data
df = pd.read_csv(r'raw_data.csv')
df
```

```python
# Sort data
df = df.sort_values(by='web-scraper-order')
df
```

```python
# Drop unnecessary columns
df.drop('web-scraper-order', inplace=True, axis=1)
df.drop('web-scraper-start-url', inplace=True, axis=1)
df.drop('letter', inplace=True, axis=1)
df.drop('letter-href', inplace=True, axis=1)
df.drop('title', inplace=True, axis=1)
df
```

```python
import re

ascii_character = r'[^\x00-\x7F]+'

# Remove non-ascii characters from a string
def remove_non_ascii(text):
    sub_answer = re.sub(ascii_character,' ', text)
    sub_answer = ''.join(filter(whitelist.__contains__, sub_answer))
    pat = re.compile(r'\s+')
    return pat.sub(' ', sub_answer)

# Test
whitelist = set("abcdefghijklmnopqrstuvwxyz ABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890'.")
myStr = "happy t00 go 129.129    $%^&*(\u201ckkkk''       '''''"
answer = remove_non_ascii(myStr)
print(myStr + '\n' + answer)

re.sub(ascii_character,' ', myStr)
```

```python
# Group the dataset by the novels
novel = df['novel'].unique()
group = df.groupby(by='novel')
dataset = []

# Create a document for each novel
for n in novel:
    index = group.groups[n]
    story = ''

    for d in df['story'][index]:
        js_data = json.loads(d)
        chapter = ''
        for jsd in js_data:
            chapter += jsd['story']
        story += chapter

    story = remove_non_ascii(story)

    doc = {
        "title": re.sub(ascii_character,' ', df['novel'][index[0]]),
        "url": df['novel-href'][index[0]],
        "author": re.sub(ascii_character,' ', df['author'][index[0]]),
        "year": str(df['year'][index[0]]),
        "origin": df['origin'][index[0]],
        "genre": df['genre'][index[0]],
        "keywords": str(df['keyword'][index[0]]).split(', '),
        "story": story
    }

    dataset.append(doc)

# Write the documents into a JSON file
with open('data.json', 'w') as outfile:
    json.dump(dataset, outfile)
```

```python
# Convert the documents into ND-JSON format
index = 1
with open("data.json", "r") as read_file:
    data = json.load(read_file)
result = [json.dumps(record) for record in data]
with open('data2.json', 'w') as obj:
    for i in result:
        obj.write('{"index":{"_id":' + str(index) + '}}\n')
        index += 1
        obj.write(i+'\n')
```
